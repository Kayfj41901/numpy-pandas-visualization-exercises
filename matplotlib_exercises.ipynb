Graph y = (x**2) - (x) + 2: 

```python
x = np.linspace(-10,10,10)
y = (x**2) - (x) + 2 
plt.plot(x, y, '-r', label='y=x**2-x+2')
plt.title('Graph of y=x**2-x+2')
plt.xlabel('X AXIS', color='#1C2833')
plt.ylabel('Y AXIS', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y = sqrt(x) 

```python
#start, end, increment
x = np.linspace(0, 100, 100)
y = [math.sqrt(n) for n in x]

plt.annotate('origin', xy = (0,0), arrowprops= dict(facecolor = 'blue'))
plt.plot(x, y, 'r')
plt.title("Graph of y=x^0.5")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

Graph y = x ** 3 Power: 

```python
x = np.linspace(-100, 100, 100)
y = x ** 3
fig = plt.figure()

plt.annotate('origin', xy = (0,0), arrowprops= dict(facecolor = 'blue'))
plt.plot(x, y, 'r')
plt.title("Graph of y=x**3")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

Graph the line 2**x 

```python
x = np.linspace(-5, 5, 10)
y = 2 ** x
fig = plt.figure()

plt.annotate('origin', xy = (0,0), arrowprops= dict(facecolor = 'blue'))
plt.plot(x, y, 'r')
plt.title("Graph of y=2**x")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# 100 linearly spaced numbers 
x = np.linspace(-5, 5, 100)


# the function
y = (1)/(x+1)

#the setting the axes at the centre
fig = plt.figure()
plt.title("Graph of y = (1)/(x+1)")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.annotate('origin', xy = (0,0), arrowprops= dict(facecolor = 'blue'))
# plot the function
plt.plot(x, y, 'r')

#show the plot 
plt.show()
```

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = np.linspace(0, 100, 10)
y1 = x ** .5
y2 = x ** 3 
y3 = [(2**i) for i in x]
y4 = [(1/(i+1)) for i in x]
y = [y1, y2, y3, y4]

plt.figure()
plt.style.use('default')

for i in range(len(y)): 
    plt.subplot(2,2,i+1)
    plt.plot(x,y[i])
    plt.grid()
    
    
plt.tight_layout()
```
