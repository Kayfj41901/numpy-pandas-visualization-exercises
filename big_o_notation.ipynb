Big O Notation

```python
plt.title = "Big O Notation"
plt.xlabel= "Elements"
plt.ylabel= "Operations"

plt.figure(figsize=(20, 20))

first_x = [x for x in range(1,10)]
y1 = [1 for n in first_x]
plt.plot(first_x, y1, c='blue')

second_x = [x for x in range(1,10)]
y2 = [math.log(n) for n in second_x]
plt.plot(second_x, y2, c='orange')

third_x = [x for x in range(1,10)]
y3 = [n for n in third_x]
plt.plot(third_x, y3, c='red')

fourth_x = [x for x in range(1,10)]
y4 = [n * math.log(n) for n in fourth_x]
plt.plot(fourth_x, y4, c='green')

fifth_x = [x for x in range(1,10)]
y5 = [n**2 for n in fifth_x] 
plt.plot(fifth_x, y5, c='purple')

sixth_x = [x for x in range(1,10)]
y6 = [2**n for n in sixth_x]
plt.plot(sixth_x, y6, c='brown')

seventh_x = [x for x in range(1,10)]
y7 = [math.factorial(n) for n in seventh_x]
plt.plot(seventh_x, y7, c='black')

eighth_x = [x for x in range(1,10)]
y8 = [n**n for n in eighth_x] 
plt.plot(eighth_x, y8, c='grey')

plt.legend(loc='upper right')

plt.show()

```
